```python
import numpy as np

input_data = np.linspace(-5, 5, 10)

print(input_data)
```

```python
# 1. Sigmoid Activation Function
# Mathematical Formula: f(x) = 1 / (1 + e^-x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

output_sigmoid = sigmoid(input_data)
print("\nSigmoid Output :", output_sigmoid)
```

```python
import matplotlib.pyplot as plt

x_sigmoid = np.linspace(-10, 10, 100)
y_sigmoid = 1 / (1 + np.exp(-x_sigmoid))

plt.figure(figsize=(8, 5))
plt.plot(x_sigmoid, y_sigmoid, color='blue', linewidth=2)
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 2. Tanh (Hyperbolic Tangent) Activation Function
# Mathematical Formula: f(x) = (e^x - e^-x) / (e^x + e^-x)

def tanh(x):
    return np.tanh(x)

output_tanh = tanh(input_data)
print("\nTanh Output :", output_tanh)
```

```python
x_tanh = np.linspace(-4, 4, 100)
y_tanh = np.tanh(x_tanh)

plt.figure(figsize=(8, 5))
plt.plot(x_tanh, y_tanh, color='green', linewidth=2)
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
plt.ylim(-1.2, 1.2)
plt.show()
```

```python
# 3. ReLU (Rectified Linear Unit) Activation Function
# Mathematical Formula: f(x) = max(0, x)

def relu(x):
    return np.maximum(0, x)

output_relu = relu(input_data)
print("\nReLU Output :", output_relu)
```

```python
x_relu = np.linspace(-3, 6, 100)
y_relu = np.maximum(0, x_relu)

plt.figure(figsize=(8, 5))
plt.plot(x_relu, y_relu, color='red', linewidth=2)
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 4. Leaky ReLU Activation Function
# Mathematical Formula: f(x) = x if x > 0, else alpha * x (alpha is small : 0.01)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

output_leaky = leaky_relu(input_data)
print("\nLeaky ReLU Output :", output_leaky)
```

```python
x_leaky = np.linspace(-5, 5, 100)
alpha = 0.01
y_leaky = np.where(x_leaky > 0, x_leaky, x_leaky * alpha)

plt.figure(figsize=(8, 5))
plt.plot(x_leaky, y_leaky, color='purple', linewidth=2)
plt.title('Leaky ReLU Activation Function (alpha=0.01)')
plt.xlabel('x')
plt.ylabel('leaky_relu(x)')
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=0, color='gray', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 5. Softmax Activation Function
# Mathematical Formula: f(x)_i = e^(x_i) / sum(e^(x_j))

#since softmax is usually applied to the output layer, we take custom input here 
softmax_inputs = np.array([2.0, 1.0, 0.1])

def softmax(x):
    e_x = np.exp(x - np.max(x)) # Subtract max for numerical stability
    return e_x / e_x.sum(axis=0)

output_softmax = softmax(softmax_inputs)
print("\nSoftmax Inputs:", softmax_inputs)
print("Softmax Probabilities:", output_softmax)
```

```python
categories = ['Class A', 'Class B', 'Class C', 'Class D']
logits = np.array([3.0, 1.5, 0.5, 2.0])
e_logits = np.exp(logits - np.max(logits))
softmax_probs = e_logits / e_logits.sum()

plt.figure(figsize=(8, 5))
plt.bar(categories, softmax_probs, color=['coral', 'skyblue', 'lightgreen', 'gold'])
plt.title('Softmax Activation Function')
plt.xlabel('Classes')
plt.ylabel('Probability')
plt.ylim(0, 1)
for i, prob in enumerate(softmax_probs):
    plt.text(i, prob + 0.02, f'{prob:.3f}', ha='center')
plt.show()
```
